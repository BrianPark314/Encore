```python
path = "Data/review_2022/"
file_list = os.listdir(path)

print ("file_list: {}".format(file_list))
index = []
AvgScore = []
stdev = []
for s in file_list:
    if 'review' not in s:
        continue
    d = pd.read_csv(path+s)
    index.append(int(s.split('_')[0]))
    AvgScore.append(d.Rank.mean())
    stdev.append(np.round(d.Rank.std(), 2))

review_score_df = pd.DataFrame({'Rank':index, 'AvgScore':AvgScore, 'stdev':stdev})
review_score_df.set_index('Rank', inplace=True)
review_score_df = review_score_df.sort_index()
review_score_df.to_csv('Data/_review_score_22.csv')
```

```python
_review_score_19_df.index += 1
_review_score_19_df['viewers'] = (boxoffice_19_df.관객수)
_review_score_22_df.index += 1
_review_score_22_df['viewers'] = (boxoffice_22_df.관객수)
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=_review_score_19_df.AvgScore, y=_review_score_19_df.stdev, size=_review_score_19_df.viewers, color='blue', sizes=(40, 400), legend=False)
sns.regplot(x=_review_score_19_df.AvgScore, y=_review_score_19_df.stdev, color='blue',scatter=False, order=4, label='19년도')
sns.scatterplot(x=_review_score_22_df.AvgScore, y=_review_score_22_df.stdev, size=_review_score_19_df.viewers, legend=False,color='red', sizes=(40, 400))
sns.regplot(x=_review_score_22_df.AvgScore, y=_review_score_22_df.stdev, color='red', scatter=False, order=4, label='22년도')
#plt.legend(labels=['19년도 박스오피스', '22년도 박스오피스', '피팅', '피팅'])
plt.legend()
plt.title('박스오피스 탑 50위 영화 리뷰의 평균 점수와 표준 편차 비교', size=20)
plt.xlabel('평균 점수', size=15)
plt.ylabel('표준 편차', size=15)
plt.ylim(0, 5)
plt.tight_layout()
plt.savefig('movie_revs.png')
plt.show()
```

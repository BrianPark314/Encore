Seaborn

Seaborn은 Matplotlib을 기반으로 다양한 색상 테마와 통계 차트를 추가한 패키지이다. countplot()을 사용하면 각 카테고리 값마다 데이터 빈도가 얼마나 있는지 표시할 수 있다. set_theme()은 global 범위의 테마를 적용해서 그래프를 쉽게 꾸밀 수 있다. style과 palette 키워드 인수를 활용한다.

```python
sns.set_theme(style='whitegrid')
```

타이타닉 데이터셋으로 카운트 플롯을 적용해 보자.

```python
df = sns.load_dataset('titanic')
sns.countplot(y=df['class'])
```

hue 키워드 인수를 활용하여 분류할 두 번째 변수를 설정할 수 있다. 

```python
sns.countplot(data=df, x='class', hue='alive')
```

플롯을 수평하게도 만들 수 있다.

```python
sns.countplot(data=df, y='deck', hue='alive')
```

```python
tips = sns.load_dataset('tips')
tips.info()
```

러그 플롯은 데이터 위치를 x축 위해 작은 선분으로 나타내어 실제 데이터들의 위치를 보여준다. 커널 밀도는 커널이라는 함수를 겹치는 방법으로 히스토그램보다 부드러운 형태의 분포 곡선을 보여준다.

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
sns.rugplot(data=tips, x='total_bill', y='tip')
```

rugplot에도 hue 인수를 추가하여 입체적인 결과를 도출할 수 있다.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
sns.rugplot(data=tips, x='total_bill', y='tip', hue='time')
```

rugplot의 크기를 조정할 수 있다.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
sns.rugplot(data=tips, x='total_bill', y='tip', height=.1)
```

데이터가 많을 때는 투명도를 설정해 rug를 좀 더 명확하게 표현할 수 있다.

```python
diamonds = sns.load_dataset('diamonds')
sns.scatterplot(data=diamonds, x = 'carat', y='price', s=3)
sns.rugplot(data=diamonds, x='carat', y='price', lw=1, alpha=.005)
```

histplot()은 1차원 혹은 2차원 데이터의 분포를 히스토그램으로 보여준다. 

```python
penguins = sns.load_dataset('penguins')
penguins.info()
```

```python
sns.histplot(data=penguins, x='flipper_length_mm')
```

bin의 너비도 당연히 지정할 수 있다.

```python
sns.histplot(data=penguins, x='flipper_length_mm', binwidth=3)
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', bins=30)
```

kde 함수를 겹쳐서 그릴 수도 있따.

```python
sns.histplot(data=penguins, x='flipper_length_mm', kde=True)
```

hue 키워드로 종을 분류하면 좀 더 명확해진다.

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='species')
```

step 인자로 오버래핑된 그래프를 잘 표현할 수 있따.

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', element='step')
```

다른 모양을 선택할 수 있다.

```python
sns.histplot(data=penguins, x='flipper_length_mm', hue='species', element='poly')
```

x와 y축에 모두 값을 할당하면 heatmap 형태로 나타낸다.

```python
sns.histplot(data=penguins, x='bill_depth_mm', y='body_mass_g', hue='species')
```

컬럼 중 하나가 이산된 값이라면 표현하기가 훨씬 쉽다.

```python
sns.histplot(penguins, x='bill_depth_mm', y='species', hue='species', legend=False)
```

displot()으로도 히스토그램을 그릴 수 있다.

```python
sns.displot(data=penguins, x='flipper_length_mm')
```

```python
sns.displot(data=penguins, x='flipper_length_mm', kind='kde')
```

누적 분포 그래프를 그려보자.

```python
sns.displot(data=penguins, x='flipper_length_mm', kind='ecdf')
```

kde를 오버랩하고 싶다면 kde 인수를 True로 설정하면 된다.

```python
sns.displot(data=penguins, x='flipper_length_mm', kde=True)
```

히트맵도 생성할 수 있다.

```python
sns.displot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
```

여기서도 kde와 rug를 적용할 수 있다.

```python
sns.displot(data=penguins, x='flipper_length_mm', y='bill_length_mm', kind='kde', rug=True)
```

hue와 kde를 동시에 사용하면 아래와 같이 된다.

```python
sns.displot(data=penguins, x='flipper_length_mm', hue='species', kind='kde')
```

col 인수를 사용해서 subset을 한번 더 분류할 수 있다.

```python
sns.displot(data=penguins, x='flipper_length_mm', hue='species', col='sex', kind='kde')
```

height와 aspect를 활용해서 그래프 크기를 제어할 수 있다. 높이 대비 너비 비율을 지정한다고 생각하면 된다.

```python
sns.displot(data=penguins, x='flipper_length_mm', hue='sex', col='species', kind='ecdf', height=4, aspect=.7)
```

각 axis label과 title도 지정해줄 수 있다.

```python
g = sns.displot(data=penguins, y='flipper_length_mm', hue='sex', col='species', kind='kde', height=4, aspect=.7)

g.set_axis_labels('Density (a.u.)', 'Flipper length (mm)')
g.set_titles("{col_name} penguins")
```

바 플롯은 카테고리마다의 평균을 각 사격형의 길이로 편차를 에러바로 표현한다. 아래를 보면 평균과 신뢰구간을 구한 것을 볼 수 있다.

```python
df = sns.load_dataset('penguins')
sns.barplot(data=df, x='species', y='body_mass_g', hue='sex')
```

boxplot은 box-whisker 플롯을 그린다.

```python
df = sns.load_dataset('titanic')
sns.boxplot(x=df.age)
```

카테고리를 추가할 수 있다.

```python
sns.boxplot(df, x='age', y='class')
```

order 키워드로 원하는 ticklabel을 전달하면 순서를 바꿀 수 있다.

```python
sns.boxplot(df, x='fare', y='alive', order=['yes', 'no'])
```

```python
sns.boxplot(data=df[['age', 'fare']], orient='v')
```

바이올린 플롯

```python
sns.violinplot(df, x='class', y='age', hue='alive')
```

```python
sns.violinplot(df, x='deck', y='age', hue='alive', split=True)
```

hue의 level이 2가 아닐 경우 split이 작동하지 않는다. 값의 표현 범위를 명확히 표현하고 싶으면 cut을 이용한다. 

```python
sns.violinplot(df, x='deck', y='age', hue='alive', cut=0)
```

대역폭을 바꾸고 싶다면 bw 인수를 바꿔준다.

```python
sns.violinplot(df, x='age', y='alive', bw=.15)
```

바이올린 내부에 분포를 표시할 수도 있따. inner 키워드를 사용하면 된다.

```python
sns.violinplot(df, x='age', y='embark_town', inner='stick', cut=0)
```

scale에 따라 각 그래프의 크기가 달라진다. 

```python
sns.violinplot(df, x='age', y='embark_town', scale='count', cut=0)
```

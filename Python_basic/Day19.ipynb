Matplotlib 개요

Matplotlib은 파이선에서 자료를 차트나 플롯으로 시각화하는 패키지이다. 뒤에서 다루게 될 seaborn의 근간이 되는 패키지이기도 하다. pyplot는 서브채키지는 matlab의 시각화 명령을 제공한다. 아래와 같이 호출하는 것이 관례이다.

플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 하고 틱 라벨을 붙일 수 있다.

```python
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 256)
C = np.cos(x)
plt.title('x축과 y축의 tick label 설정')
plt.plot(x, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

틱 라벨 문자열에는 '$$' 사이에 수학 문자식을 넣을 수도 있다.

```python
x = np.linspace(-np.pi, np.pi, 256)
C = np.cos(x)
plt.title('x축과 y축의 tick label 설정')
plt.plot(x, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```

여러 선을 한꺼번에 그리고 싶은 경우 다음과 같이 실행한다.

```python
t = np.arange(0., 5., 0.2)
plt.title('라인 플롯에서 여러 개의 선 그리기')
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
plt.show()
```

범례(Legend)를 추가할 수 있다. 위치는 자동으로 결정되지만 loc 인수를 통해 위치를 변경할 수 있다.

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.title('범례를 표시한 플롯')
plt.plot(x, c, ls='--', label='cosine')
plt.plot(x, s, ls=':', label='sine')
plt.legend(loc=2)
plt.show()
```

figure 안에 axes를 생성하려면 원래 subplot()을 사용하여 명시적으로 axes 객체를 얻어야 한다. 그러나 plot()으로 그려도 된다. 위/아래 두 개의 플롯을 하나의 피규어 안에 그릴려면 다음과 같이 해야 한다. 여기서 숫자 인덱싱은 파이썬이 아닌 Matlab 관행을 따르기 때문에 첫 번째 플롯을 가리키는 숫자가 0이 아니라 1임에 주의해야 한다. tight_layout()명령을 실행하면 플롯 간의 간격을 자동으로 맞춰준다.

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1) *np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped Oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

subplots()에 인수를 두개만 전달하면 복수의 axes 객체를 동시에 생성할 수도 있다.

```python
fig, axes = plt.subplots(2, 2)
np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title('axes 1')
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title('axes 2')
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title('axes 3')
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title('axes 4')

plt.tight_layout()
plt.show()
```

xerr 인수나 yerr인수를 지정하면 에러 바를 추가할 수 있다. alpha는 투명도이다.

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
performance = 3+10*np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title('Barh Chart')
plt.barh(people, performance, xerr=error, alpha=0.4)
plt.xlabel('x 라벨')
plt.show()
```

파이 차트

카테고리 별 값의 상대적인 비교를 해야 할 때는 pie 명령으로 파이 차트를 그릴 수 있다. 

```python
labels = ['월세','전세','자가','기타']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0, 0.1, 0)
plt.title('Pie chart')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal') #원의 형태 유지
plt.show()
```

연습문제

```python
import pandas as pd
titanic = sns.load_dataset('titanic')
```

```python
bins = [0, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]

#titanic.dropna(inplace=True)
titanic_cats = pd.cut(titanic.age, bins, labels=labels)
tcats=  titanic_cats.value_counts(normalize=True)

plt.title('타이타닉호 승객 비율')
plt.pie(tcats, explode=explode, labels=tcats.index, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
